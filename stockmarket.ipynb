```python
import os
import pandas as pd
flag=0 #buy
flags=[]
df=pd.read_csv("Downloads\ACC_with_indicators_.csv")
df
```

```python
for i in range(len(df)):
    if df.loc[i,'close'] < df.loc[i,'lowerband']:
        flag=0
    if df.loc[i,'close'] > df.loc[i,'upperband']:
        flag=1
    if df.loc[i,'RSI8'] > 70 and df.loc[i,'RSI14'] > 70:
        flag=1
    if df.loc[i,'fastk'] > df.loc[i,'fastd']:
        flag=0
    if df.loc[i,'RSI8'] < 30 and df.loc[i,'RSI14'] < 30:
        flag=0
    macd=['macd510','macd520','macd1020','macd1226']
    macd_df=df.iloc[i,df.columns.get_indexer(macd)]
    lar=macd_df.max()
    if lar > 0:
        flag=0
    ma=['sma5','sma10','sma15','sma20','ema5','ema10','ema15','ema20']
    ma_df=df.iloc[i,df.columns.get_indexer(ma)]
    ln=ma_df.max()
    if df.loc[i,'close'] > ln:
        flag=1
    df.loc[i,'recommendation']=int(flag)
df['recommendation']
```

```python
df.to_csv('ACC_with_indicators_.csv', index=False)
```

```python
ic=['close','sma5','sma10','sma15','sma20','ema5','ema10','ema15','ema20','upperband','lowerband','macd510','macd520','macd1020','macd1520','macd1226','RSI14','RSI8','fastk','fastd']
input=df[ic].copy()
input
```

```python
output=df['recommendation']
output
```

```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets, linear_model, metrics
X_train, X_test,y_train, y_test = train_test_split(input,output,test_size=0.4,random_state=1)
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
pred=reg.score(X_test, y_test)
pred
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
k = 5 
knn = KNeighborsClassifier(n_neighbors=k)


knn.fit(X_train_scaled, y_train)

y_pred = knn.predict(X_test_scaled)


accuracy = accuracy_score(y_test, y_pred)
accuracy
```
